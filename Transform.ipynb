```python
def transform_brand(x):
    if isinstance(x, str) and 'Marque:' in x:
        return x.replace("Marque:", "")
    else:
        return "Autre"
```

```python
df['Brand']=df['Brand'].apply(transform_brand)
```

```python
df['Price'] = pd.to_numeric(df['Price'].str.replace('Dhs', '', regex=False)
                                      .str.replace(',', ''), errors='coerce')
df['Old Price'] = pd.to_numeric(df['Old Price'].str.replace('Dhs', '', regex=False)
                                                .str.replace(',', ''), errors='coerce')
df['Discount']=df['Discount'].apply(lambda x:x.replace('%',''))
df['Price_Diff']=df['Old Price']-df['Price']
df['Rating']=df['Rating'].apply(lambda x:x.replace('out of 5',''))
df
```

```python
def categorize_stock(x):
    if x=='Nan' or x=='NaN':
        return "No Info"
    elif x=="Disponible":
        return "High Stock"
    else:
        return "Low Stock"
df['Stock']=df['Stock'].apply(categorize_stock)
```

```python
def extract_number(text):
    # Use regex to find digits in the string
    match = re.search(r'\d+', str(text))
    if match:
        return int(match.group())
    else:
        return 0

df['Reviews']=df['Reviews'].apply(extract_number)
```

```python
def extract_seller_info(info):
    # Use regex to find the evaluation percentage and the number of subscribers
    evaluation_match = re.search(r'(\d+)%', info)  # Extracts percentage
    subscribers_match = re.search(r'(\d+) Abonnés', info)  # Extracts subscriber number
    
    evaluation = int(evaluation_match.group(1)) if evaluation_match else 0  # Default to 0 if not found
    subscribers = int(subscribers_match.group(1)) if subscribers_match else 0  # Default to 0 if not found
    
    return pd.Series([evaluation, subscribers])
```

```python
df[['Seller_Evaluation', 'Seller_Abonne_Number']] = df['Seller Info'].apply(extract_seller_info)
```

```python
def extract_performance_info(info):
    # Use regex to find the relevant values
    shipping_speed_match = re.search(r'Vitesse d\'expédition:\s*(\w+\s\w+|\w+)', info)  # Extracts shipping speed
    quality_score_match = re.search(r'Qualité:\s*(\w+\s\w+|\w+)', info)  # Extracts quality score
    consumer_review_match = re.search(r'Avis des consommateurs:\s*(\w+\s\w+|\w+)', info)  # Extracts consumer review

    shipping_speed = shipping_speed_match.group(1).replace('Qualité','') if shipping_speed_match else "neutre"  # Default to "N/A" if not found
    quality_score = quality_score_match.group(1).split('Avis')[0] if quality_score_match else "neutre"  # Default to "N/A" if not found
    consumer_review = consumer_review_match.group(1) if consumer_review_match else "neutre"  # Default to "N/A" if not found

    return pd.Series([shipping_speed, quality_score, consumer_review])
```

```python
df[['Shipping_Speed', 'Quality_Score', 'Consumer_Review']] = df['Seller Performance'].apply(extract_performance_info)
```

```python

def get_sentiment_score(text):
    """
    Analyzes the sentiment of the given text and returns the polarity score.
    
    Parameters:
    text (str): The input text for sentiment analysis.
    
    Returns:
    float: The sentiment polarity score, ranging from -1 (negative) to 1 (positive).
    """
    blob = TextBlob(text)
    sentiment_score = blob.sentiment.polarity
    return sentiment_score

```

```python
df['First Comment']=df['First Comment'].apply(get_sentiment_score)
```

```python
df['Second Comment']=df['Second Comment'].apply(get_sentiment_score)
```

```python
def label_encode(column):
    """Encodes a categorical column using Label Encoding."""
    unique_values = column.unique()
    encoding_dict = {value: index for index, value in enumerate(unique_values)}
    encoded_column = column.map(encoding_dict)
    return encoded_column
```

```python
from sklearn.cluster import KMeans
import joblib
def apply_kmeans(df, n_clusters=3):
    """
    Applies K-means clustering to a DataFrame and returns it with a 'Cluster' column.

    Parameters:
    df (pd.DataFrame): Input DataFrame with numeric features.
    n_clusters (int): Number of clusters for K-means (default is 3).

    Returns:
    pd.DataFrame: DataFrame with an additional 'Cluster' column.
    """
    # Initialize K-means model
    kmeans = joblib.load('kmeans_model.pkl')

    # Fit the model and predict cluster labels
    df['Cluster'] = kmeans.predict(df)
    joblib.dump(kmeans,'kmeans_model.pkl')
    return df
```
